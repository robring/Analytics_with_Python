## Data Correction

```python
df['state'] = df['state'].astype('category')
df['state']
```

```python
cat_convert = ['emp_title', 'homeownership', 'verification_income_joint']

for i in cat_convert:
    df[i] = df[i].astype("category")
```

```python
df['metric'] = df['annual_income'] / df['total_credit_limit']
```
