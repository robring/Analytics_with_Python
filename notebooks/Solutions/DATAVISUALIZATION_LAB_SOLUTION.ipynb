# Scatter Plots in Matplot
Create a scatter plot of the iris data set using matplot. Implement the following steps:

- load iris dataset
- print iris feature names to know what to select later
- store sepal features in a list named sepal_features 
- sepal length shall be plotted on x axis
- sepal width shall be plotted on y axis
- target value shall be used for color
- size (parameter s) shall be calculated as 5*sum of petal features
- define and show the scatter plot, choose alpha=1 to have identical colors wihtin a class

```python
import matplotlib.pyplot as plt
from sklearn import datasets

#load iris dataset
iris = datasets.load_iris()
features, target = iris.data, iris.target

#print iris feature names to know what to select later
print(iris.feature_names)

#store sepal features in a list named sepal_features 
sepal_features =features[:,0:2]

#sepal length shall be plotted on x axis
x = sepal_features[:,0]
#sepal width shall be plotted on y axis
y = sepal_features[:,1]
#target value shall be used for color
colors = target
#size (parameter s) shall be calculated as 5*sum of petal features
size = (features[:,2]+features[:,3])*5

#define and show the scatter plot, choose alpha=1 to have identical colors wihtin a class
plt.scatter(x, y, s=size, c=colors, alpha=1)
plt.show()
```
