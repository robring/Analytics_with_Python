- Load wine dataset from sklearn.datasets
- Print feature names
- Store the alcohol and color_intensity (CI) values for each wine example in a new 2-dim array named al_co
- Print al_co values

```python
import sklearn.datasets as ds
data = ds.load_wine()
print(data.feature_names)
al_co = data.data[:,[0,9]]
print(al_co)
```

Plot a simple boxplot of the CI values using the matplotlib.pyplot package

```python
import matplotlib.pyplot as plt

plt.boxplot(al_co)
plt.show()
```

Use an adequate feature scaling function to bring all values into the range [0,1]

Hint: you will need the preprocessing package

- Fit the scaler
- Transform the al_co values using the fitted scaler
- Print and boxplot the fitted values

```python
from sklearn.preprocessing import MinMaxScaler

minmax_scaler = MinMaxScaler()
minmax_scaler.fit(al_co)

scaled_al_co=minmax_scaler.transform(al_co)
print(scaled_al_co)

plt.boxplot(scaled_al_co)
plt.show()
```

Now use a different scaler to rescale the al_co values to have a specific mean (=0) and a specific standard deviation (=1)

Hint: you will need the preprocessing package

- Fit the scaler
- Transform the al_co values using the fitted scaler
- Boxplot the fitted values

```python
from sklearn.preprocessing import StandardScaler

std_scaler = StandardScaler()
std_scaler.fit(al_co)

scaled_al_co=std_scaler.transform(al_co)


plt.boxplot(scaled_al_co)
plt.show()
```

Concatenate an outlier example ([20,20]) to al_co
- Boxplot the concatenated data

Use RobustScaler to rescale the al_co values 

- Fit the scaler
- Transform the al_co values using the fitted scaler
- Boxplot the fitted values

```python
import numpy as np
from sklearn.preprocessing import RobustScaler

al_co = data.data[:,[0,9]]
print(type(al_co))
al_co=np.concatenate((al_co,[[20,20]]))

plt.boxplot(al_co)
plt.show()

rob_scaler = RobustScaler()
rob_scaler.fit(al_co)

scaled_al_co=rob_scaler.transform(al_co)

plt.boxplot(scaled_al_co)
plt.show()
```
